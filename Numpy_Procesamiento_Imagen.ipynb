## Importar la librerías 

```python
#from PIL import Image  #Pillow
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Leer una imagen un arreglo

```python
img = mpimg.imread('lenaColor.png')

print(img.dtype)
print(img.ndim)
print(img.shape)
```

##  Desplegar una imagen  que es un arreglo

```python
plt.imshow(img)
plt.show()
```

## Recortar una imagen con slice

```python
im_trim1 = img[128:384, 128:384]
print(im_trim1.shape)
plt.imshow(im_trim1)
plt.show()
```

## Generación de un solo color y concatenación de imágenes

```python
im_R = img.copy()
im_R[:, :, (1, 2)] = 0 #cambiar a 1
plt.imshow(im_R)
plt.show()
```

```python
im_G = img.copy()
im_G[:, :, (0, 2)] = 0
plt.imshow(im_G)
plt.show()
```

```python
im_B = img.copy()
im_B[:, :, (0, 1)] = 0
plt.imshow(im_B)
plt.show()
```

```python
im_RGB = np.concatenate((im_R, im_G, im_B), axis=1)
plt.imshow(im_RGB)
plt.show()
```

## Reducción de color

```python
from PIL import Image  #Pillow
im = np.array(Image.open('lenaColor.png').resize((256, 256)))

#im_32 = im // 32 * 32
im_128 = im // 128 * 128
plt.imshow(im_128)
plt.show()
```

```python
im90=np.rot90(im)
plt.imshow(im90)
plt.show()
```

## Rotaciones

```python
#imf=np.fliplr(im)
imf=np.flipud(im)
plt.imshow(imf)
plt.show()
```
